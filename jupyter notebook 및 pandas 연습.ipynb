##### Data Frame methods

속성과 달리 메소드는 괄호를 포함한다.
모든 속성과 메소드는 dir 함수로 나열할 수 있다.

- head(n), tail(n) : 첫번째와 마지막의 n개의 행을 얻음
- describe() : 설명을 생성?
- max() , min() : 해당 열의 최대값과 최소값을 리턴함.
- mean() , median() : 숫자로된 열의 값의 중간값을 반환
- std() : 표준편차 출력
- sample(n) : 데이터 프레임의 무작위 샘플을 반환
- dropna() : 값이 없는 레코드를 삭제
